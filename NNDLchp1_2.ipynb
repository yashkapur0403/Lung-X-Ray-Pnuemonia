```python
import numpy as np
import random

class Network(object):
  def __init__(self, sizes):
    self.num_layers = len(sizes)     #3 layers total
    self.sizes = sizes               # network store karta layer-wise kitne neuron
    self.weights = [np.random.randn (y,x) for x, y in zip(sizes[:-1], sizes[1:])]
    self.bias = [np.random.randn (y,1) for y in sizes[1:]]

  def feedforward (self, a):
    for w, b in zip(self.weights, self.bias):      # weights and bias ko b, w named variables me daal diya
      a = self.sigmoid(np.dot(w,a)+b)            #applying sigmoid on the neuron
    return a


  def SGD(self, training_data, mini_batch_size, epoch, eta, test_data):
    if test_data:
      n_test = len(test_data)
    n = len(training_data)

    for j in range(epoch):
      random.shuffle(training_data)

      mini_batches = [training_data[k : k + mini_batch_size] for k in range(0, n, mini_batch_size)]     # minibatches creation

      for mini_batch in mini_batches:               #update minibatch function ko call kiya hai
        self.update_mini_batch(mini_batch, eta)

      if test_data:        #if we have test data by chance

        print(f"Epoch {j} : {self.evaluate(test_data)} / {n_test}")
      else:
        print(f"Epoch {j} completed")


  def update_mini_batch(self, mini_batch, eta):
    nabela_b = [np.zeros(b.shape) for b in self.bias]         #initialse gradients of b,w to 0
    nabela_w = [np.zeros(w.shape) for w in self.weights]

    for x, y in mini_batch:
      delta_nabela_b, delta_nabela_w = self.backprop(x,y)          #backprop function call

      nabela_b = [nb + dnb for nb, dnb in zip(nabela_b, delta_nabela_b)]
      nabela_w = [nw + dnw for nw, dnw in zip(nabela_w, delta_nabela_w)]      #add each updated training eg to gradient

    self.bias = [b -(eta/len(mini_batch))* nb for b, nb in zip(self.bias, nabela_b)]          #updating
    self.weights = [w - (eta/len(mini_batch))*nw for w, nw in zip(self.weights, nabela_w)]


  def backprop(self, x, y):
    nabela_b = [np.zeros(b.shape) for b in self.bias]
    nabela_w = [np.zeros(w.shape) for w in self.weights]

    activation = x
    activations = [x]
    zs = []

    for b, w in zip(self.bias, self.weights):
        z = np.dot(w, activation) + b
        zs.append(z)
        activation = self.sigmoid(z)
        activations.append(activation)

    # error for outer layer using eq-1 of backprop
    delta = self.cost_derivative(activations[-1], y) * self.sigmoid_prime(zs[-1])
    nabela_b [-1] = delta
    nabela_w [-1] = np.dot(delta, activations[-2].T)

    # error for hidden layers using eq-2
    for l in range (2, self.num_layers):
      z = zs[-l]
      sp = self.sigmoid_prime(z)

      delta = np.dot(self.weights[-l+1].T, delta)*sp
      nabela_b [-l] = delta
      nabela_w [-l] = np.dot(delta, activations[-l-1].T)

    return (nabela_b, nabela_w)

  def cost_derivative(self, output_activations, y):
      return (output_activations - y)

  def evaluate(self, test_data):
      test_results = [(np.argmax(self.feedforward(x)), y) for (x, y) in test_data]
      return sum(int(x == y) for (x, y) in test_results)

  def sigmoid(self, z):
    return 1.0/(1.0+ np.exp(-z))

  def sigmoid_prime(self, z):
    return self.sigmoid(z) * (1 - self.sigmoid(z))



from google.colab import files
files.upload()

training_data, validation_data, test_data = load_data_wrapper()

net = Network([784, 30, 10])
net.SGD(training_data, 30, 10, 3.0, test_data=test_data)
```
